## Line Chart

```python
# Plotting a line chart df

from matplotlib import pyplot as plt
import pandas as pd

df_sales = pd.read_excel('linechart.xlsx')
df_sales.head(5)


# plt.figure(figsize=(12,4)) most common convection
plt.figure(figsize=(12,4))


# plt.plot(x-axis, y-axis) line chart uses both axis
plt.plot(df_sales['Quarter'], df_sales['Fridge'], label='Fridge')
plt.plot(df_sales['Quarter'], df_sales['Dishwasher'], label='Dishwasher')
plt.plot(df_sales['Quarter'], df_sales['Washing Machine'], label='Washing Machine')

# adding title to the chart
plt.title("Product Sales")

# adding names to x and y axis
plt.xlabel("Quarter")
plt.ylabel("Million $")

# showing each slaes colorwise
plt.legend()

plt.show()
```

## Pie Chart

```python
total_sales = df_sales[["Fridge", "Dishwasher", "Washing Machine"]].sum()
total_sales
```

```python
# labels = for putting their names on pie chart
# autopct = for putting percentage on pie chart
# startangel = for rounding the pie chart
# explode = (0.1,1,1) for extracting a particulart pie form chart
# shadow = for making more attractive

plt.pie(total_sales, labels=total_sales.index, autopct='%1.1f%%', startangle=140, explode=(0.1,0,0), shadow=True)
plt.show()
```

# Histogram

```python
plt.hist(df_score['Exam_Score'], density = True)
# for histogram seaborn is much efficent
```
